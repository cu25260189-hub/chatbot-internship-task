```python
def chatbot():
    print("ChatBot: Hey! I'm your friendly chatbot.")
    print("ChatBot: Type 'bye' anytime to end the chat.\n")

    while True:
        user_input = input("You: ").lower().strip()

        if user_input == "hi" or user_input == "hello":
            print("ChatBot: Hello! Nice to meet you.")

        elif "how are you" in user_input:
            print("ChatBot: I'm doing great. Thanks for asking. How about you?")

        elif "your name" in user_input:
            print("ChatBot: I'm ChatBot, created using Python.")

        elif "help" in user_input:
            print("ChatBot: Sure. You can ask me about my name, how I am, or just chat.")

        elif "thank" in user_input:
            print("ChatBot: You're welcome. Happy to help.")

        elif user_input == "bye":
            print("ChatBot: Goodbye. Have a great day.")
            break

        else:
            print("ChatBot: I didn't understand that. Can you try saying it differently?")


# Run the chatbot
chatbot()
```
